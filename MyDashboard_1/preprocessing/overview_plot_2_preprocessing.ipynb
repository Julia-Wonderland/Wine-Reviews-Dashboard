```python
import numpy as np
import pandas as pd
import os
```

```python
parent = os.path.dirname(os.getcwd())
parent
```

```python
wine_data = pd.read_csv(parent + '/data/wine_df_emb_nona_nosam.csv', sep=',')
wine_data.head()
```

```python
wine_data["embedding"]
```

```python
wine_data.drop(columns = ["embedding", "pca_x", "pca_y", "tsne_x", "tsne_y"], inplace=True)
```

```python
wine_data.drop(columns = ["Unnamed: 0", "region_1", "region_2", "taster_name", 
                          "taster_twitter_handle", "title"], inplace=True)
wine_data.head()
```

```python
grouped_wine_data = wine_data.groupby(['winery', 'variety']).agg(
        {"points" : "mean", "price" : "mean", "country" : "count"}).reset_index(
        ).rename(columns = {"country" : "amount_of_wines"}).round(2)
grouped_wine_data.head()
```

```python
grouped_wine_data["description"] = wine_data["description"]
```

```python
grouped_wine_data.to_csv(parent + '/data/grouped_wine_df_nona.csv', sep=',', index = False)
test_wine_data = pd.read_csv(parent + '/data/grouped_wine_df_nona.csv', sep=',')
test_wine_data.head()
```
