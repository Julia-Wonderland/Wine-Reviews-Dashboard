```python
import numpy as np
import pandas as pd
import os
```

```python
parent = os.path.dirname(os.getcwd())
parent
```

```python
wine_data = pd.read_csv(parent + '/data/wine_df_emb_nona_nosam.csv', sep=',')
wine_data.head()
```

```python
wine_data.drop(columns = ["embedding", "pca_x", "pca_y", "tsne_x", "tsne_y"], inplace=True)
wine_data.drop(columns = ["Unnamed: 0", "country", "title", "description"], inplace = True)
wine_data.head()
```

```python
wine_data["taster_name"].value_counts()
```

```python
reviewers_data = wine_data.groupby("taster_name").agg({"points" : "mean", "price" : "mean", 
                                                       "variety" : "count"}).reset_index()
reviewers_data.rename(columns = {"taster_name" : "reviewer", "variety" : "num_reviews"}, inplace = True)
reviewers_data.set_index("reviewer", inplace = True)
reviewers_data.head()
```

```python
best_variety = wine_data.groupby(["taster_name", "variety"]).agg({"points" : "mean"}).reset_index()
best_variety = best_variety.groupby("taster_name").apply(lambda x: x.loc[x['points'].idxmax()]).reset_index(drop=True)
best_variety.set_index("taster_name", inplace = True)
best_variety.head()
```

```python
best_winery = wine_data.groupby(["taster_name", "winery"]).agg({"points" : "mean"}).reset_index()
best_winery = best_winery.groupby("taster_name").apply(lambda x: x.loc[x['points'].idxmax()]).reset_index(drop=True)
best_winery.set_index("taster_name", inplace = True)
best_winery.head()
```

```python
best_province = wine_data.groupby(["taster_name", "province"]).agg({"points" : "mean"}).reset_index()
best_province = best_province.groupby("taster_name").apply(lambda x: x.loc[x['points'].idxmax()]).reset_index(drop=True)
best_province.set_index("taster_name", inplace = True)
best_province.head()
```

```python
reviewers_data["best_variety"] = best_variety["variety"]
reviewers_data["best_winery"] = best_winery["winery"]
reviewers_data["best_province"] = best_province["province"]
reviewers_data.head()
```

```python
reviewers_data["points"] = reviewers_data["points"].round(2)
reviewers_data["price"] = reviewers_data["price"].round(2)
reviewers_data = reviewers_data.T
reviewers_data.index.name = "attributes"
reviewers_data.to_csv(parent + '/data/reviewers_aggregated_data.csv', sep=',')
reviewers_data.head()
```
